Aunque herramientas ms nuevas como ggplot y ggvis en el lenguaje R, junto con kits de herramientas de visualizacin web basados en D3js y HTML5 canvas, o incluso libreras python como seaborn o plotly a menudo hacen que Matplotlib parezca un poco anticuado, su fortaleza como motor grfico probado y multiplataforma no puede ser ignorada.

### Dos interfaces por el precio de uno

Una caracterstica potencialmente confusa de Matplotlib es sus dos interfaces: una conveniente interfaz basada en el estado al estilo de MATLAB, y una interfaz orientada a objetos ms poderosa. Destacaremos rpidamente las diferencias entre las dos aqu.
```


#### Interfaz al estilo MATLAB (Como lo vers en muchos sitios)

**Matplotlib fue originalmente escrito como una alternativa de Python para usuarios de MATLAB**, y gran parte de su sintaxis refleja ese hecho.
Las herramientas al estilo MATLAB estn contenidas en la interfaz pyplot (``plt``).


````python
fig = plt.figure()

plt.subplot(2, 2, 1) # filas, columnas, indice  4 subplots 
plt.plot(x, np.sin(x))


plt.subplot(2, 2, 2) # filas, columnas, indice
plt.plot(x, np.cos(x));


plt.subplot(2, 2, 3) # filas, columnas, indice
plt.plot(x, 2*x*x/100);

plt.subplot(2, 2, 1) # filas, columnas, indice
plt.plot(x, 2*x*x/100);

#todo es una figura pero tiene 4 subplots
````

Es importante tener en cuenta que esta interfaz es *stateful*: lleva un registro de la **"figura" y ejes "actuales", en los cuales se aplican todos los comandos de ``plt``.**
Puedes obtener una referencia a estos utilizando las rutinas ``plt.gcf()`` (obtener figura actual) y ``plt.gca()`` (obtener ejes actuales).

Si bien esta forma de actuar de la interfaz es rpida y conveniente para grficos simples, es fcil encontrarse con problemas.
Por ejemplo, una vez que se crea el segundo panel, cmo podemos retroceder y agregar algo al primero?
Esto es posible dentro de la interfaz de estilo MATLAB, pero es un poco engorroso.
Afortunadamente, hay una mejor manera.

````python
plt.subplot(2,2,1)
plt.plot(x,np.full(shape=(100,),fill_value= 1))
````

````python
display(fig)
````


#### Interfaz orientada a objetos

La interfaz orientada a objetos est disponible para situaciones ms complicadas y cuando se desea tener ms control sobre la figura.
En lugar de depender de alguna nocin de una figura o ejes "activos", en la interfaz orientada a objetos las funciones de trazado son *mtodos* de objetos ``Figure`` y ``Axes`` explcitos.
Para volver a crear el grfico anterior usando este estilo de trazado, podras hacer lo siguiente:


````python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x, np.sin(x))
ax[1,1].plot(x, np.cos(x))
ax[1,1].plot(x, 0*x) #le aado algo a la 1,1 que es diagonal dcha abajo
ax[0,1].plot(x, x)
#la 1,0 esta vaca
````

````python
ax[0,1].plot(x, x-3)
````

````python
display(fig)
````

````python
fig.savefig('data/prueba.png')
````

````python
Image('data/prueba.png')
````


Para grficos ms simples, la eleccin del estilo a utilizar es en gran medida una **cuestin de preferencia, pero el enfoque orientado a objetos puede convertirse en una necesidad a medida que los grficos se vuelven ms complicados.**  

A lo largo de esta unidad, alternaremos entre las interfaces de estilo MATLAB y orientado a objetos, dependiendo de lo que sea ms conveniente.  

En la mayora de los casos, la diferencia es tan pequea como cambiar ``plt.plot()`` a ``ax.plot()``, pero hay algunas cosas a tener en cuenta que destacaremos a medida que surjan en las siguientes sesiones.
